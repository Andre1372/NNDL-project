## Test Rete Neurale Semplice

Creiamo una rete neurale molto semplice per verificare che PyTorch funzioni correttamente.

```python
import torch.nn as nn

# Definizione rete semplice
class SimpleNet(nn.Module):
    def __init__(self):
        super(SimpleNet, self).__init__()
        self.fc1 = nn.Linear(10, 5)
        self.fc2 = nn.Linear(5, 2)
        self.relu = nn.ReLU()
    
    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.fc2(x)
        return x

# Creazione modello
model = SimpleNet()
print("Architettura del modello:")
print(model)

# Test forward pass
x_test = torch.randn(1, 10)
output = model(x_test)
print(f"\nInput shape: {x_test.shape}")
print(f"Output shape: {output.shape}")
print(f"Output: {output}")
```
